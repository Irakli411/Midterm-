```python
import pandas as pd
import zipfile
import numpy as np
```

```python
zip_path = r"C:\Users\Hp\Desktop\74300220.zip"

with zipfile.ZipFile(zip_path) as z:
    with z.open("data_ws2025/ALB.csv") as f:
        ALB = pd.read_csv(f)
        ALB['Ticker'] = 'ALB'
        ALB['Date'] = pd.to_datetime(ALB['Date']) 
    with z.open("data_ws2025/IP.csv") as f:
        IP = pd.read_csv(f)
        IP['Ticker'] = 'IP'
        IP['Date'] = pd.to_datetime(IP['Date']) 
    with z.open("data_ws2025/NWSA.csv") as f:
        NWSA = pd.read_csv(f)
        NWSA['Ticker'] = 'NWSA'
        NWSA['Date'] = pd.to_datetime(NWSA['Date']) 
    with z.open("data_ws2025/ORLY.csv") as f:
        ORLY = pd.read_csv(f)
        ORLY['Ticker'] = 'ORLY'
        ORLY['Date'] = pd.to_datetime(ORLY['Date']) 
    with z.open("data_ws2025/TXT.csv") as f:
        TXT = pd.read_csv(f)
        TXT['Ticker'] = 'TXT'
        TXT['Date'] = pd.to_datetime(TXT['Date']) 

TXT.dtypes

    
```

```python
data = pd.concat([ALB, IP, NWSA, ORLY, TXT], ignore_index=True)
data
```

```python
#Part 1


data[data['High'] == data['Low']]
# There is no such company, meaninng their stocks changes during each day for all of them
```

```python
#Part 2
data.groupby('Ticker').agg({'High': 'max', 'Low': 'min'})
```

```python
#Part 3
data['log_close'] = np.log(data['Close']) 

returns = data.groupby('Ticker')['log_close'].agg(['min', 'max', 'mean', 'std', 'skew'])

returns
```

```python
#part 4 
data['diff'] = data['Open'] - data['Close']
```

```python
#max
data_sorted = data.sort_values(['Ticker', 'diff'], ascending=[True, False])
data_sorted['rank'] = data_sorted.groupby('Ticker').cumcount()
data_sorted[data_sorted['rank'] == 1][['Ticker', 'diff']]

```

```python
#min
data_sorted = data.sort_values(['Ticker', 'diff'], ascending=[True, True])
data_sorted['rank'] = data_sorted.groupby('Ticker').cumcount()
data_sorted[data_sorted['rank'] == 1][['Ticker', 'diff']]


```
